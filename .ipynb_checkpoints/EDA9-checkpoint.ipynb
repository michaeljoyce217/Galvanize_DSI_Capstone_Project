```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('data/crime_csv_all_years.csv',parse_dates={'dttime':[1,2,3]}, keep_date_col=True)
```

```python
df['day_of_week']=df['dttime'].dt.weekday_name
```

```python
df_temp=df.copy()
```

```python
df_temp.head()
```

```python
df2=df_temp.dropna()
#remove missing data, all (or nearly all) of which is the non-property crime data
```

```python
df3=df2.rename(index=str, columns={"YEAR": "year", "MONTH": "month", "DAY":"day","HOUR":"hour", "MINUTE":"minute", "NEIGHBOURHOOD":"neighborhood"})
#rename columns as all caps is annoying
```

```python
df4=df3.sort_values(['year','month','day','hour','minute'])
#sort by date
```

```python
df5=df4.drop(['minute', 'HUNDRED_BLOCK','TYPE'], axis=1)
#remove extraneous data 
```

```python
df6=df5.apply(pd.to_numeric, errors='ignore')
```

```python
#bin by 12am-8am, 8am-4pm, 4pm -12am
hourbins = [-0.1,8.0,16.0,24.1]
hourlabels = ['1200am-0759am', '0800am-0359pm', '0400pm-1159pm']
df6['day_segment'] = pd.cut(df6["hour"], bins=hourbins,labels=hourlabels)

```

```python
df7=df6[['year', 'month', 'day', 'day_segment', 'neighborhood']]
#group by neighborhood, by day_segment

```

```python
df8=df7.groupby(df7.columns.tolist()).size()
```

```python
df9=pd.DataFrame(df8).reset_index()
```

```python
df10=df9.rename(index=str, columns={ 0 :"number_of_crimes"})
```

```python
df_final=df10.copy()
```

```python
wdf3=wdf2.rename(index=str, columns={ " YEARMODA":"date","PRCP": "prcp", "TMAX":"tmax","TMIN":"tmin"})
```

```python
def get_year(x):
    return int (str(x)[0:4])

def get_month(x):
    return int (str(x)[4:6])

def get_day(x):
    return int(str(x)[6:8])

def get_prcp(x):
    return float(x[1:5])
    
get_year(20030101)
```

```python
wdf3['year'] = wdf3['date'].apply(get_year)
wdf3['month'] =  wdf3['date'].apply(get_month)
wdf3['day'] = wdf3['date'].apply(get_day)
wdf3['precipitation']=wdf3['prcp'].apply(get_prcp)
```

```python
wdf3.head()
```

```python
wdf4=wdf3[['year','month','day','tmax', 'tmin', 'precipitation']]
```

```python
wdf_final=wdf4.copy()
```

```python
new_df1=pd.merge(wdf_final,df_final, how='left', on=['year','month','day'])
```

```python
X=new_df_final3[['year', 'month', 'day', 'tmax', 'tmin', 'precipitation',
     'consumer_price_index', 'gdp_millions_2007',
       'seasonally_adjusted_unemployment', 'unadjusted_unemployment',
       'Possession, cocaine ', 'Heroin, possession ',
       'neighborhood_Central Business District',
       'neighborhood_Dunbar-Southlands', 'neighborhood_Fairview',
       'neighborhood_Grandview-Woodland', 'neighborhood_Hastings-Sunrise',
       'neighborhood_Kensington-Cedar Cottage', 'neighborhood_Kerrisdale',
       'neighborhood_Killarney', 'neighborhood_Kitsilano',
       'neighborhood_Marpole', 'neighborhood_Mount Pleasant',
       'neighborhood_Musqueam', 'neighborhood_Oakridge',
       'neighborhood_Renfrew-Collingwood', 'neighborhood_Riley Park',
       'neighborhood_Shaughnessy', 'neighborhood_South Cambie',
       'neighborhood_Stanley Park', 'neighborhood_Strathcona',
       'neighborhood_Sunset', 'neighborhood_Victoria-Fraserview',
       'neighborhood_West End', 'neighborhood_West Point Grey',
       'day_segment_0800am-0359pm', 'day_segment_0400pm-1159pm']]

y=new_df_final3['number_of_crimes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)  # Don't cheat - fit only on training data
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)  # apply same transformation to test data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df=pd.DataFrame(lm.coef_, X.columns, columns=['Coefficeint'])
coeff_df
```

```python
from sklearn import metrics
predictions=lm.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
